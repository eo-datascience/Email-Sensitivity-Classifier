```python
silver_data = spark.read.parquet(
    "/Volumes/workspace/default/email_warehouse/silver/clean_emails"
)
```

```python
silver_data.printSchema()
```

```python
silver_data.count()
```

```python
top_senders = silver_data.groupBy("from_email") \
    .count() \
    .orderBy("count", ascending=False)
top_senders.display()
```

```python
from pyspark.sql.functions import countDistinct

sender_unique_contacts = silver_data.groupBy("from_email") \
    .agg(countDistinct("to_email").alias("unique_contacts")) \
    .orderBy("unique_contacts", ascending=False)

sender_unique_contacts.display()
```

```python
from pyspark.sql.functions import countDistinct, count
#combine volume and breadth of emails
sender_stats = silver_data.groupBy("from_email") \
    .agg(
        count("from_email").alias("total_emails"),
        countDistinct("to_email").alias("unique_contacts")
    ) \
    .orderBy("total_emails", ascending=False)

sender_stats.display()
```
